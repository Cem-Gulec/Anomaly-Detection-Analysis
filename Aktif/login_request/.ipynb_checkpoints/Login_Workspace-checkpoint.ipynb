## Metrics used

- Mean absolute error (MAE) is a measure of errors between paired observations expressing the same phenomenon. <br>
$MAE = \frac{\sum\limits_{i=1}^{n} |y_i - \hat{y}_i|}{n}$

- The mean absolute percentage error (MAPE), is a measure of prediction accuracy of a forecasting method in statistics. <br>
$MAPE = \frac{100}{n}\sum\limits_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{y_i}$ 

```python
# MAPE metric
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Calculate average of last n observations
# A moving average (rolling average or running average) is a calculation to analyze data points 
# by creating a series of averages of different subsets of the full data set.
def moving_average(series, n):
    return np.average(series[-n:])
```

```python
# series: actual values
# window: rolling window size 
# plot_intervals: plot confidence intervals
# plot_anomalies: plot anomalies 
def plot_moving_average(series, window, plot_intervals=False, plot_anomalies=False):

    scale_value = 1.95
    # data created by rolling a windows with "window" size
    rolling_mean = series.rolling(window = window).mean()
    # error value calculated with MAPE metric
    error = mean_absolute_percentage_error(series.values, rolling_mean[:len(series)])

    plt.figure(figsize=(15,7))
    plt.title(f"Moving average at window size = {window}")
    plt.plot(rolling_mean, "g", label = "Rolling mean trend, error {:.2f}%".format(error[0]))

    # Plot confidence intervals 
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bound = rolling_mean - (mae + scale_value * deviation)
        upper_bond = rolling_mean + (mae + scale_value * deviation)
        plt.plot(upper_bond, "r--", label="Upper / Lower Bond")
        plt.plot(lower_bound, "r--")
        
        # Find anomaly related values 
        if plot_anomalies:
            anomalies = pd.DataFrame(index = series.index, columns = series.columns)
            # values under the lower bound
            anomalies[series < lower_bound] = series[series < lower_bound]
            # value above the upper bound
            anomalies[series > upper_bond] = series[series > upper_bond]
            
            plt.plot(anomalies, "ro", markersize = 8)
        
    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="upper left")
    plt.grid(True)
```

```python
plot_moving_average(login_df, 4, plot_intervals=True, plot_anomalies=True) 
```

## Exponential Smoothing
Exponential smoothing is a rule of thumb technique for smoothing time series data using the exponential window function. Whereas in the simple moving average the past observations are weighted equally, exponential functions are used to assign exponentially decreasing weights over time. <br>
$\hat{y}_{t} = \alpha \cdot y_t + (1-\alpha) \cdot \hat y_{t-1} $ <br>

## Double Exponential Smoothing
Simple exponential smoothing does not do well when there is a trend in the data. In such situations, several methods were devised under the name "double exponential smoothing" or "second-order exponential smoothing," which is the recursive application of an exponential filter twice. <br>
The basic idea behind double exponential smoothing is to introduce a term to take into account the possibility of a series exhibiting some form of trend. This slope component is itself updated via exponential smoothing.

$\ell_x = \alpha y_x + (1-\alpha)(\ell_{x-1} + b_{x-1})$

$b_x = \beta(\ell_x - \ell_{x-1}) + (1-\beta)b_{x-1}$

$\hat{y}_{x+1} = \ell_x + b_x$

```python
# series: actual values
# alpha: smoothing parameter ranging from [0.0, 1.0]
def exponential_smoothing(series, alpha):
    
    result = [series[0]] 
    
    # iterate over remaining values but initial one
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    
    return result

# series: timeseries dataset
# alpha: smoothing parameter for level
# beta: smoothing parametere for trend
def double_exponential_smoothing(series, alpha, beta):
    
    result = [series[0]]
    
    for n in range(1, len(series) + 1):
        if n == 1:
            level, trend = series[0], series[1] - series[0]
        if n >= len(series): 
            value = result[-1]
        else:
            value = series[n]
        last_level, level = level, alpha * value + (1 - alpha) * (level + trend)
        trend = beta * (level - last_level) + (1 - beta) * trend
        result.append(level + trend)
        
    return result
```

```python
# Plots both single and double exponential smoothing with different alphas and betas
# single exponential smoothing only requires "alphas" parameter, but double needs both "alphas" and "betas" 
# series: actual values
# alphas: smoothing parameters for level
# betas: smoothing parameters for trend
def plot_exponential_smoothing(series, alphas, betas = None):
    
    with plt.style.context('seaborn-white'):    
        plt.figure(figsize=(15, 7))
        
        # Double exponential smoothing
        if betas:
            for alpha in alphas:
                for beta in betas:
                    error = mean_absolute_percentage_error(series.values, double_exponential_smoothing(series, alpha, beta)[:len(series)])
                    plt.plot(double_exponential_smoothing(series, alpha, beta), label = "Alpha {}, beta {}, error {:.2f}%".format(alpha, beta, error))
        
        # Single exponential smoothing
        else:
            for alpha in alphas:
                error = mean_absolute_percentage_error(series.values, exponential_smoothing(series, alpha)[:len(series)])
                plt.plot(exponential_smoothing(series, alpha), label="Alpha {}, error {:.2f}%".format(alpha, error))
                
        plt.plot(series.values, "c", label = "Actual")
        plt.legend(loc="best")
        plt.axis('tight')
        
        if betas:
            plt.title("Double Exponential Smoothing")
        else:
            plt.title("Single Exponential Smoothing")
            
        plt.grid(True)
```

## Triple Exponential Smoothing (HoltWinters)
The Holt-Winters ES modifies the Holt ES technique so that it can be used in the presence of both trend and seasonality.

```python
# Triple exponential smoothing which is also called Holtwinters
from statsmodels.tsa.holtwinters import ExponentialSmoothing

actual_values = login_df.RequestResponseLogDuration.values

# Trend: Additive, Seasonal: Additive
login_df["Holt_Add"] = ExponentialSmoothing(login_df.RequestResponseLogDuration, trend="add", seasonal="add", seasonal_periods=24).fit().fittedvalues
error1 = mean_absolute_percentage_error(actual_values, login_df["Holt_Add"][:len(actual_values)])

# Trend: Additive, Seasonal: Multiplicative
login_df["Holt_AddMul"] = ExponentialSmoothing(login_df.RequestResponseLogDuration, trend="add", seasonal="mul", seasonal_periods=24).fit().fittedvalues
error2 = mean_absolute_percentage_error(actual_values, login_df["Holt_AddMul"][:len(actual_values)])

# Trend: Multiplicative, Seasonal: Multiplicative
login_df["Holt_Mult"] = ExponentialSmoothing(login_df.RequestResponseLogDuration, trend="mul", seasonal="mul", seasonal_periods=24).fit().fittedvalues
error3 = mean_absolute_percentage_error(actual_values, login_df["Holt_Mult"][:len(actual_values)])

# Trend: Multiplicative, Seasonal: Additive
login_df["Holt_MultAdd"] = ExponentialSmoothing(login_df.RequestResponseLogDuration, trend="mul", seasonal="add", seasonal_periods=24).fit().fittedvalues
error4 = mean_absolute_percentage_error(actual_values, login_df["Holt_MultAdd"][:len(actual_values)])

fig, axs = plt.subplots(2, 2, figsize=(20, 7))
plt.figure(figsize=(20,10))

axs[0, 0].plot(login_df["Holt_Add"])
axs[0, 0].set_title("Holt_Add, error {:.2f}%".format(error1))

axs[0, 1].plot(login_df["Holt_AddMul"])
axs[0, 1].set_title("Holt_AddMul, error {:.2f}%".format(error2))

axs[1, 0].plot(login_df["Holt_Mult"])
axs[1, 0].set_title("Holt_Mult, error {:.2f}%".format(error3))

axs[1, 1].plot(login_df["Holt_MultAdd"])
axs[1, 1].set_title("Holt_MultAdd, error {:.2f}%".format(error3))
```

## Applying lags to time series data
Datasetin n birim kaydırılıp (t - n) modelin üretilen bu yapay datasetle eğitilmesi teorik olarak ileri tahminlemede kullanılabilecek bir yöntemdir. <br>
Fakat bunun kötü yanı şudur ki, geride bir yerde elimizdeki featurelardan bağımsız sert değişimler gerçekleşirse model bunu yakalayamayacaktır. <br>
Dolayısıyla farklı lag aralıkları optimal pencereyi sunabilir ve problemin komple çözümünde kullanılmasa dahi parça parça bu verilerden faydalanılabilir.

```python
# Split into train and test sets with corresponding test_size parameter provided
def split_train_test(X, y, test_size):
    
    # get the index after which test set starts
    test_index = int(len(X) * (1-test_size))
    
    X_train = X.iloc[:test_index]
    y_train = y.iloc[:test_index]
    X_test = X.iloc[test_index:]
    y_test = y.iloc[test_index:]
    
    return X_train, X_test, y_train, y_test
```

```python
# Plots modelled vs fact values, prediction intervals and anomalies
def plot_model_results(model, X_train = X_train, X_test = X_test, plot_intervals=False, plot_anomalies=False):
    
    scale = 1.95
    prediction = model.predict(X_test)
    
    plt.figure(figsize=(15, 7))
    plt.plot(prediction, "g", label="prediction", linewidth=2.0)
    plt.plot(y_test.values, label="actual", linewidth=2.0)
    
    if plot_intervals:
        cv = cross_val_score(model, X_train, y_train, cv=tscv, scoring = "neg_mean_absolute_error")
        mae = cv.mean() * (-1)
        deviation = cv.std()
        
        lower = prediction - (mae + scale * deviation)
        upper = prediction + (mae + scale * deviation)
        
        plt.plot(lower, "r--", label = "upper bond / lower bond", alpha = 0.5)
        plt.plot(upper, "r--", alpha = 0.5)
        
        if plot_anomalies:
            anomalies = np.array([np.NaN] * len(y_test))
            anomalies[y_test < lower] = y_test[y_test < lower]
            anomalies[y_test > upper] = y_test[y_test > upper]
            plt.plot(anomalies, "o", markersize=10, label = "Anomalies")
    
    error = mean_absolute_percentage_error(prediction, y_test)
    plt.title("Mean absolute percentage error {0:.2f}%".format(error))
    plt.legend(loc="best")
    plt.tight_layout()
    plt.grid(True);
```

```python
data = pd.DataFrame(login_df.RequestResponseLogDuration.copy())
data.columns = ["y"]
```

```python
# Adding lag feature columns to the dataframe with a range of [3,31]
for i in range(3, 31):
    data["lag_{}".format(i)] = data.y.shift(i)
```

```python
y = data.dropna().y
X = data.dropna().drop(['y'], axis=1)

# 30% of data is reserved for testing
X_train, X_test, y_train, y_test = split_train_test(X, y, test_size = 0.3)
```

## Trying out new features alongside with lag features created

```python
data.index = pd.to_datetime(data.index)
data["hour"] = data.index.hour
data["weekday"] = data.index.weekday
data['is_weekend'] = data.weekday.isin([5,6])*1
data.tail()
```

```python
# In order to enforce all of the categories to be in the same value type
# values are scaled under the same range.
scaler = StandardScaler()
y = data.dropna().y
X = data.dropna().drop(['y'], axis=1)

# split dataset
X_train, X_test, y_train, y_test = split_train_test(X, y, test_size = 0.3)

# scale both train and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# train linear regression model
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

# Plot results from linear regression model
plot_model_results(lr, X_train=X_train_scaled, X_test=X_test_scaled, plot_intervals=True, plot_anomalies=True)
```

```python
# series:          actual values
# lag_start:       initial lag step number 
#                  lag_start = 1 would mean predict today from values of 1 day before
# lag_end:         final lag step number
#                  lag_end = 30 would mean predict today from values of 1 month before
# test_size:       split size
# target_encoding: add target averages to the dataset
def prepare_data(series, lag_start, lag_end, test_size, target_encoding = False):
    
    # copy of the initial dataset
    data = pd.DataFrame(series.copy())
    data.columns = ["y"]
    
    # lags of series
    for i in range(lag_start, lag_end):
        data["lag_{}".format(i)] = data.y.shift(i)
    
    # datetime features
    data.index = pd.to_datetime(data.index)
    data["hour"] = data.index.hour
    data["weekday"] = data.index.weekday
    data['is_weekend'] = data.weekday.isin([5,6])*1
    
    if target_encoding:
        # calculate averages on just the train set
        test_index = int(len(data.dropna())*(1-test_size))
        
        # keys: unique categories of the specified feature (e.g. weekday)
        # values: means over real feature (y in this case)
        code_mean_weekday = dict(data[:test_index].groupby('weekday')["y"].mean())
        code_mean_hour = dict(data[:test_index].groupby('hour')["y"].mean())
        
        data['weekday_average'] = list(map(code_mean_weekday.get, data.weekday))
        data["hour_average"] = list(map(code_mean_hour.get, data.hour))

        # frop encoded variables 
        data.drop(["hour", "weekday"], axis=1, inplace=True)
    
    # train-test split
    y = data.dropna().y
    X = data.dropna().drop(['y'], axis=1)
    X_train, X_test, y_train, y_test = split_train_test(X, y, test_size=test_size)

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = prepare_data(login_df.RequestResponseLogDuration, lag_start = 3, lag_end = 31, test_size = 0.3, target_encoding = True)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

plot_model_results(lr, X_train=X_train_scaled, X_test=X_test_scaled, plot_intervals=True, plot_anomalies=True)
```
